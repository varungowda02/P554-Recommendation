```python
rating_with_name=ratings.merge(books,on='ISBN')
```

```python
num_rating_df=rating_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
df=round(num_rating_df.merge(avg_rating_df,on='Book-Title'),2)
df
```

```python
df.num_rating.value_counts()
```

```python
df=df[df['num_rating']>=250].sort_values('avg_rating',ascending=False)
df
```

```python
df=df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
df
```

```python
# collaborative filtering recommendation system
```

```python
x=rating_with_name.groupby('User-ID')['Book-Rating'].count()>200
educated_user=x[x].index

# "These are the users who rated the book more than 200"
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(educated_user)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title')['Book-Rating'].count()>=50
y[y]

# these sre the books which get rated more than 50
```

```python
famous_book=y[y].index
famous_book
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
final_rating
```

```python
final_rating.duplicated().sum()
```

```python
book_user_pivot=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
book_user_pivot.fillna(0,inplace=True)
```

```python
book_user_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(book_user_pivot)
print(similarity_score.shape)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(book_user_pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == book_user_pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
